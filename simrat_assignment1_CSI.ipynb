Q- Create lower triangular, upper triangular and pyramid containing the "*" character.

```python
# LOWER TRIANGLE
def lower_tri(n):
    for i in range(1, n+1):
        print('*' * i)

lower_tri(5)
```

```python
# UPPER TRIANGLE
def upper_tri(n):
    for i in range(n+1,0,-1):
        print('*' * i)

upper_tri(5)
```

```python
# PYRAMID
def pyramid(n):
    for i in range(n+1):
        print(' ' * (n-i) + '*' * (2*i-1))

pyramid(5)
```

# EXTRA INPUT

Here i have created a menu-driven Python program that allows users to select from a variety of patterns (including left- right triangles, diamonds,butterfly, number, and alphabet patterns), customize the size and symbol, choose alignment and even preview or save the output into a txt file.

```python
import sys

def get_int(prompt, min_value=1, max_value=100):

    while True:
        try:
            value = int(input(prompt))
            if value < min_value or value > max_value:
                print(f"Please enter a value between {min_value} and {max_value}.")
            else:
                return value
        except ValueError:
            print("Invalid input. Please enter a valid integer.")

def get_symbol():
    while True:
        symbol = input("Enter a symbol, digit, or alphabet (single character): ")
        if len(symbol) == 1:
            return symbol
        print("Please enter exactly one character.")

def get_alignment():

    while True:
        align = input("Choose alignment: (L)eft, (R)ight, (C)enter: ").lower()
        if align in ('l', 'r', 'c'):
            return align
        print("Invalid choice. Please enter L, R, or C.")

def preview_and_save(pattern_lines):

    print("\nPattern Preview:\n")
    for line in pattern_lines:
        print(line)
    save = input("\nSave pattern to a file? (y/n): ").lower()
    if save == 'y':
        filename = input("Enter filename (e.g., pattern.txt): ")
        try:
            with open(filename, 'w') as f:
                for line in pattern_lines:
                    f.write(line + '\n')
            print(f"Pattern saved to {filename}")
        except Exception as e:
            print(f"Error saving file: {e}")

def lower_triangle(n, symbol, align):

    lines = []
    for i in range(1, n+1):
        line = symbol * i
        if align == 'r':
            line = line.rjust(n)
        elif align == 'c':
            line = line.center(n)
        lines.append(line)
    return lines

def upper_triangle(n, symbol, align):
    lines = []
    for i in range(n, 0, -1):
        line = symbol * i
        if align == 'r':
            line = line.rjust(n)
        elif align == 'c':
            line = line.center(n)
        lines.append(line)
    return lines

def pyramid(n, symbol, align):

    lines = []
    for i in range(1, n+1):
        line = symbol * (2*i-1)
        if align == 'l':
            line = line.ljust(2*n-1)
        elif align == 'r':
            line = line.rjust(2*n-1)
        else:
            line = line.center(2*n-1)
        lines.append(line)
    return lines

def diamond(n, symbol, align):

    lines = []
    for i in range(1, n+1):
        line = symbol * (2*i-1)
        if align == 'l':
            line = line.ljust(2*n-1)
        elif align == 'r':
            line = line.rjust(2*n-1)
        else:
            line = line.center(2*n-1)
        lines.append(line)
    for i in range(n-1, 0, -1):
        line = symbol * (2*i-1)
        if align == 'l':
            line = line.ljust(2*n-1)
        elif align == 'r':
            line = line.rjust(2*n-1)
        else:
            line = line.center(2*n-1)
        lines.append(line)
    return lines

def hourglass(n, symbol, align):

    lines = []
    for i in range(n, 0, -1):
        line = symbol * (2*i-1)
        if align == 'l':
            line = line.ljust(2*n-1)
        elif align == 'r':
            line = line.rjust(2*n-1)
        else:
            line = line.center(2*n-1)
        lines.append(line)
    for i in range(2, n+1):
        line = symbol * (2*i-1)
        if align == 'l':
            line = line.ljust(2*n-1)
        elif align == 'r':
            line = line.rjust(2*n-1)
        else:
            line = line.center(2*n-1)
        lines.append(line)
    return lines

def butterfly(n, symbol, align):

    lines = []
    for i in range(1, n+1):
        left = symbol * i
        spaces = ' ' * (2*(n-i))
        right = symbol * i
        line = left + spaces + right
        if align == 'c':
            line = line.center(2*n)
        lines.append(line)
    for i in range(n, 0, -1):
        left = symbol * i
        spaces = ' ' * (2*(n-i))
        right = symbol * i
        line = left + spaces + right
        if align == 'c':
            line = line.center(2*n)
        lines.append(line)
    return lines

def number_pyramid(n, align):

    lines = []
    for i in range(1, n+1):
        nums = ''.join(str(j) for j in range(1, 2*i))
        if align == 'l':
            line = nums.ljust(2*n-1)
        elif align == 'r':
            line = nums.rjust(2*n-1)
        else:
            line = nums.center(2*n-1)
        lines.append(line)
    return lines

def alphabet_pyramid(n, align):

    lines = []
    for i in range(1, n+1):
        chars = ''.join(chr(64+j) for j in range(1, 2*i))
        if align == 'l':
            line = chars.ljust(2*n-1)
        elif align == 'r':
            line = chars.rjust(2*n-1)
        else:
            line = chars.center(2*n-1)
        lines.append(line)
    return lines

def menu():

    print("\nPATTERN GENERATOR MENU")
    print("1. Lower Triangle")
    print("2. Upper Triangle")
    print("3. Pyramid")
    print("4. Diamond")
    print("5. Butterfly")
    print("6. Number Pyramid")
    print("7. Alphabet Pyramid")
    print("0. Exit")
    return get_int("Choose a pattern (0-7): ", 0, 10)

def main():

    while True:
        choice = menu()
        if choice == 0:
            print("Thank you for using the Pattern Generator!")
            sys.exit(0)
        n = get_int("Enter number of rows (2-20): ", 2, 20)
        align = get_alignment()
        symbol = '*'
        if choice not in (9, 10):
            symbol = get_symbol()
        if choice == 1:
            lines = lower_triangle(n, symbol, align)
        elif choice == 2:
            lines = upper_triangle(n, symbol, align)
        elif choice == 3:
            lines = pyramid(n, symbol, align)
        elif choice == 4:
            lines = diamond(n, symbol, align)
        elif choice == 5:
            lines = butterfly(n, symbol, align)
        elif choice == 6:
            lines = number_pyramid(n, align)
        elif choice == 7:
            lines = alphabet_pyramid(n, align)
        else:
            print("Invalid choice.")
            continue
        preview_and_save(lines)

if __name__ == "__main__":
    main()
```
